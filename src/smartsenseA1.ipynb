```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
```

```python
df = pd.read_csv('EPI_batch/recipe_cuisine_recipe_info0.csv')
df.head(30)
```

```python
D = pd.DataFrame(columns = ['rating', 'cuisine'])
path = 'EPI_batch/'
files = os.listdir(path)

frames = []
for file in files:
    if((file != '.DS_Store') and (file != '.ipynb_checkpoints')):
      df = pd.read_csv(path+str(file))
      frames.append(df)
result = pd.concat(frames)
```

```python
print(result.columns)
print(result['rating'])
result['rating'] = [str(i) for i in result['rating']]
result['rating'] = [float(i.split("/")[0]) for i in result['rating']]
print(result['rating'])
print(result['recipe_category'])
```

```python
D['rating'] = result['rating']
D['cuisine'] = result['recipe_cuisine']
print(D)
```

```python
print(D['cuisine'].value_counts())
cuisines = list(D['cuisine'])
count = list(D['cuisine'].value_counts())
```

```python
values = D['cuisine'].value_counts()
trace = go.Bar(y=values.index[::-1],x=(values)[::-1],orientation = 'h')
layout = go.Layout(title = "Recipes per cuisine",xaxis=dict(title='Frequency of cuisine'),
                   yaxis=dict(title='Cuisine Type'))
data = [trace]
fig = go.Figure(data=data, layout=layout)
iplot(fig,filename='basic-bar')
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.model_selection import train_test_split as tts
from sklearn.model_selection import KFold, cross_val_score
import numpy as np
```

```python
for i in range(len(D['rating'])):
  # print(list(D['rating'])[i])
  if list(D['rating'])[i] =='NaN':
    list(D['rating'])[i] = 0
for i in range(len(D['cuisine'])):
  if list(D['cuisine'])[i] =='NaN':
    list(D['cuisine'])[i] = 0

X = np.array(D['rating'])
y = np.array(D['cuisine'])

X_train, X_test, y_train, y_test = tts(X, y, shuffle=True, test_size = 0.25)
model = LR(random_state=0)
model.fit(X_train.reshape(-1,1), y_train.reshape(-1,1))
```

```python
kf = KFold(n_splits=15, random_state=0)
accuracy = cross_val_score(model,X,y, cv= kf, scoring='accuracy')
print(f'The mean accuracy for the model is {accuracy.mean()} with a Std. Dev of {accuracy.std()}')
```
